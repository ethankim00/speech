## How Many Minutes of Audio Data?

```python
def load_audio(file: str):
    audio, sr = librosa.load(file, sr = 16000)
    return audio, sr 

def get_wav_files(path: str):
    filenames = []
    for subdir, dirs, files in os.walk(path):
        for file in files:
            filenames.append(os.path.join(subdir, file))
    return filenames

def get_audio_duration(audio, sr):
    return len(audio)/sr

def get_total_audio_duration(files: List, sr= 16000):
    duration = 0
    for file in files:
        audio, sr = load_audio(file)
        duration += get_audio_duration(audio, sr)
    return duration/60
```

```python
ssd_files = get_wav_files("../data/data/ssd/wav/")
ssd_duration = get_total_audio_duration(ssd_files)
print("SSD duration is:", ssd_duration)
```

```python
td_files = get_wav_files("../data/data/td/wav/")
td_duration = get_total_audio_duration(td_files)
print("SSD duration is:", td_duration)
```
